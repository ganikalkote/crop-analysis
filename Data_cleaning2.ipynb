```python
# CORRELATION

def find_corr_for_all_columns():
    print("\ncorrelation coefficients for all numeric columns\n")
    print(data.corr(method='pearson'))
find_corr_for_all_columns()
```

```python
#CORRELATION

def corr_1961_and_2014():
    corr_coeff = data['Y1961'].corr(data['Y2014'])
    print("\ncorrelation coeff between Y1961 and Y2014 is ", corr_coeff)
    if (corr_coeff == 0):
        print("Y1961 and Y2014 are not related")
    elif (corr_coeff < 0.8 and corr_coeff > 0):
        print("Year 1961 and 2014 values are directly related")
    else:
        print("Year 1961 and 2014 values are Indirectly related") 
corr_1961_and_2014()
```
